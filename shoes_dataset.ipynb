```python
import pandas as pd
import zipfile

za = zipfile.ZipFile('/content/shoes_dataset.zip')
za.namelist()

df = pd.read_csv(za.open('MEN_SHOES.csv'))
df.head()
```

```python
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
df
```

```python
for i in range(len(df)):
  df.How_Many_Sold[i] = df.How_Many_Sold[i].replace(',',"")
  df.Current_Price[i] = df.Current_Price[i].replace('₹',"")
  df.Current_Price[i] = df.Current_Price[i].replace(',',"")
```

```python
df.info()
```

```python
df.Brand_Name = df.Brand_Name.astype('str')
df.How_Many_Sold = df.How_Many_Sold.astype('int')
df.Current_Price = df.Current_Price.astype('int')
df.RATING = df.RATING.astype('float')
```

```python
df.info()
```

```python
# solving for average shoe price by brand
df.groupby('Brand_Name').mean()['Current_Price']
```

```python
# what brand has the highest prices?
df.sort_values(by=['Current_Price'],ascending=False).head()
```
